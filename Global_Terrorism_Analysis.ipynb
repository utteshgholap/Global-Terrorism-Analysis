````python
import pandas as pd
import plotly.express as px

# 1. Reset the index so 'region_txt' becomes a normal column
df_plot = Regional_terror.reset_index()

# 2. "Melt" the dataframe from wide format to long format for Plotly
df_long = df_plot.melt(
    id_vars=['region_txt'],
    var_name='Group Name',
    value_name='Incident Count'
).dropna()

# Filter out combinations with zero incidents
df_long = df_long[df_long['Incident Count'] > 0]

# 3. ADVANCED CLEANING: Identify the Top 10 most active groups globally
# and group the remaining hundreds of smaller factions into an "Other" category
top_10_groups = df_long.groupby('Group Name')['Incident Count'].sum().nlargest(10).index
df_long['Group Name'] = df_long['Group Name'].apply(lambda x: x if x in top_10_groups else 'Other')

# Re-aggregate the data now that smaller groups are combined into "Other"
df_long = df_long.groupby(['region_txt', 'Group Name'], as_index=False)['Incident Count'].sum()

# 4. SORTING: Order the regions by total number of attacks so the chart reads beautifully from highest to lowest
region_order = df_long.groupby('region_txt')['Incident Count'].sum().sort_values(ascending=True).index

# 5. Create the interactive horizontal stacked bar chart
fig = px.bar(
    df_long,
    x='Incident Count',
    y='region_txt',
    color='Group Name',
    orientation='h',
    category_orders={'region_txt': list(region_order)}, # Applies the sorted region order
    title='Regional Distribution of Terrorist Attacks by Militant Group',
    color_discrete_sequence=px.colors.qualitative.Safe # Clean, professional categorical colors
)

# 6. Fine-tune layout and design aesthetics
fig.update_layout(
    xaxis_title='Number of Attacks',
    yaxis_title='Region',
    legend_title_text='Terrorist Group',
    barmode='stack',                  # Stacks the groups per region side-by-side
    template='plotly_white',          # Crisp white presentation template
    height=700,
    margin=dict(l=200),               # Extra padding on the left so region names aren't clipped
    legend=dict(traceorder='normal')  # Normalizes legend ordering
)

# Render the interactive dashboard
fig.show()
````
